```python
import matplotlib.pyplot as plt
import numpy as np 
plt.figure(0)
method = ["Full-Single","Full-Compound","RL-DQN","ISRM"]
runtime = [255,200,204,217]
plt.bar(method,runtime)
for a,b in zip(method,runtime):
    plt.text(a,b,
             '%.0f'%b,
             ha='center', 
             va='bottom',
            )
plt.ylabel("Average latency costs/ ms")
plt.title('latency costs on optimal ')
plt.savefig("2.jpg")

plt.show()
```
